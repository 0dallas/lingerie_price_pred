## Procesamiento de data

### Importar Librerias

```python
from eccd_datasets import load_lingerie
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

datasets = load_lingerie() # use all the datasets together
```

```python
type(datasets)
```

### Imprimir los datos por cada vendedor

```python
dataset_list = []
for i in datasets:
    data_i = datasets[i]
    dataset_list.append(data_i)
    print(i.upper())
    print(data_i.shape)
    display(data_i.head(1))
    print("----------------------------------------")
DATA = pd.concat(dataset_list)
```

```python
DATA.info()
```

### Procesando columna `price`

```python
DATA['price'] = DATA['price'].str.replace("USD","")
DATA['price'] = DATA['price'].str.replace("$","")
DATA['price'] = pd.to_numeric(DATA['price'],errors='coerce')
```

### Valore nulos por columna

```python
DF_NULOS = pd.DataFrame(DATA.isna().sum(),columns=["NULOS"])
DF_NULOS['%'] = (DF_NULOS['NULOS']/613143)*100
DF_NULOS
```

```python
DATA = DATA.drop(columns=['rating','review_count','style_attributes'])
DATA = DATA.dropna()
```

### Cantidad de valores por cada columna categorica

```python
cat = list(DATA.select_dtypes(include=["object"]).columns)
num = list(DATA.select_dtypes(include=["number"]).columns)
```

```python
DATA = DATA.reset_index(drop=True)
print(DATA.shape)
for i in cat:
    print(f"---------{i}---------")
    print(DATA[i].nunique())
    valores = pd.DataFrame(DATA[i].value_counts()).rename(columns={'count':'Cantidad'})
    valores['%'] =  (valores['Cantidad']/613143)*100
    valores['Acumulado'] = valores['%'].cumsum()
    display(valores)
```

### Selección de columnas para el modelo

En base a hay muchas columnas de alta variabilidad, se eligen columnas que podría considerarse más apotantes al modelo

```python
DATA_MODELO = DATA[['price','product_category','brand_name','color']] # Sin 'product_name'
```

```python
DATA_MODELO.head(2)
```

**Columna: "brand_name"**

```python
DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].str.upper()
DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].str.replace('-',' ')
DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].str.replace("'",'')
DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].str.replace(" ",'')
DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].str.replace(".",'')

def procesar_brand(row):
    if 'VICTORIASSECRET' in row:
        return "Victoria's Secret"
    elif 'HANKYPANKY' in row:
        return "Hanky-Panky"
    elif 'WACOAL' in row:
        return "Wacoal"
    elif 'CALVINKLEIN' in row:
        return "Calvin Klein" ############# Hasta aca la mayoria
    elif 'AERIE' in row:
        return "AERIE"
    elif 'BTEMPTD' in row:
        return "B. Tempt'd"
    elif 'VANITYFAIR' in row:
        return "Vanity Fair"
    elif 'NORDSTROMLINGERIE' in row:
        return "Nordstrom"
    else:
        return "Otro"

DATA_MODELO['brand_name'] = DATA_MODELO['brand_name'].apply(procesar_brand)
```

```python
valores = pd.DataFrame(DATA_MODELO['brand_name'].value_counts()).rename(columns={'count':'Cantidad'})
valores['%'] =  (valores['Cantidad']/613143)*100
valores['Acumulado'] = valores['%'].cumsum()
# valores[valores['Acumulado']<=80]
valores
```

**Columna: "product_category"**

```python
DATA_MODELO_2 = DATA_MODELO.copy()
def categorias(d):
    if 'Bra' in d:
        return 'Bra'
    elif 'Panti' in d:
        return 'Panties'
    else:
        return d

DATA_MODELO_2['product_category_2'] = DATA_MODELO_2['product_category'].apply(categorias)
```

```python
DATA_MODELO_2[DATA_MODELO_2['product_category_2']=='Bra']['price'].value_counts()
```

```python
valores = pd.DataFrame(DATA_MODELO['product_category'].value_counts()).rename(columns={'count':'Cantidad'})
valores['%'] =  (valores['Cantidad']/613143)*100
valores['Acumulado'] = valores['%'].cumsum()
porc = 70
print(valores[valores['Acumulado']<=porc].shape)
# valores[valores['Acumulado']<=porc].join(marca)
valores = valores[valores['Acumulado']<=porc]
valores
```

*el 70% de los datos lo acupan 50 categorías por ello se van  a mantener estas y las demas se agruparan en una genérica*

```python
def recategorizar(c):
    if c in valores.index.to_list():
        return c
    else:
        return "Otro"
    
DATA_MODELO['product_category'] = DATA_MODELO['product_category'].apply(recategorizar)
```

Columna: "color"

```python
DATA_MODELO['color'] = DATA_MODELO['color'].str.lower()
DATA_MODELO['color'] = DATA_MODELO['color'].str.replace("-","")
DATA_MODELO['color'] = DATA_MODELO['color'].str.replace(" ","")
# DATA_MODELO['color'] = DATA_MODELO['color'].str.capitalize()
```

```python
DATA_MODELO_3 = DATA_MODELO.copy()
def color(d):
    if 'black' in d:
        return 'Black'
    elif 'white' in d:
        return 'White'
    elif 'ensign' in d:
        return 'Ensign'
    else:
        return d

DATA_MODELO_3['color_2'] = DATA_MODELO_3['color'].apply(color)
DATA_MODELO_3[DATA_MODELO_3['color_2']=='Black']['price'].value_counts()
```

*Hay alta variabilidad en los **precios** de productos "Black" por ello no es recomendable agrupar a los colore que llevan el nombre "black" en una sola*

```python
valores = pd.DataFrame(DATA_MODELO['color'].value_counts()).rename(columns={'count':'Cantidad'})
valores['%'] =  (valores['Cantidad']/613143)*100
valores['Acumulado'] = valores['%'].cumsum()
valores = valores[valores['Acumulado']<=50]
print(valores.shape)
valores.head()
```

```python
def colores_resumen(c):
    if c in valores.index.to_list():
        return c
    else:
        return "Otro"
    
DATA_MODELO['color'] = DATA_MODELO['color'].apply(colores_resumen)
```
